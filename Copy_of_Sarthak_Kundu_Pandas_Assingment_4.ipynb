```
#  Q1.- Make a pivot table of top 10 bowlers(in terms of wickets) with over(1-20) showing no. of wickets.  [Dificult-]
```

```

mask_1 = deliveries['dismissal_kind']=='caught'
mask_2 = deliveries['dismissal_kind']=='bowled'
mask_3 = deliveries['dismissal_kind']=='lbw'
mask_4 = deliveries['dismissal_kind']=='stumped'
mask_5 = deliveries['dismissal_kind']=='caught and bowled'
mask_6 = deliveries['dismissal_kind']=='hit wicket'
new2= deliveries[mask_1 | mask_2 |mask_3 | mask_4 |mask_5 | mask_6]





new2['bowler'].value_counts().head(10) 



mask_7 = new2['bowler']=='SL Malinga'
mask_8 = new2['bowler']=='A Mishra'
mask_9 = new2['bowler']=='Harbhajan Singh'
mask_10 = new2['bowler']=='PP Chawla'
mask_11 = new2['bowler']=='DJ Bravo'
mask_12 = new2['bowler']=='B Kumar'
mask_13 = new2['bowler']=='R Ashwin'
mask_14 = new2['bowler']=='UT Yadav'
mask_15 = new2['bowler']=='SL Malinga'
mask_16 = new2['bowler']=='RA Jadeja'
df = new2[mask_7 | mask_8 | mask_9 | mask_10 | mask_11| mask_12 | mask_13 | mask_14 | mask_15 | mask_16 ]





df.pivot_table(index='bowler',columns='over', aggfunc='count') ## per over wicket taken by the bowler
```

```
# Q2.- Make a function for batting profile of any batsman:-   [Medium+]
def profile(batsman):
  innings=new[new['batsman']==batsman].nunique()[0]
  Total_Runs=new[new['batsman']==batsman].sum()[15]
  man_of_match=matches[matches['player_of_match']==batsman].count()[0]
  fours=new[new['batsman']==batsman].groupby('batsman_runs').count().loc[4][1]
  sixes=new[new['batsman']==batsman].groupby('batsman_runs').count().iloc[5][1]
  boundaries = fours + sixes
  print('Name: ',batsman )
  print('Innings: ',innings)
  print('Total Runs: ',Total_Runs)
  print('No. of MOM(Man of the match): ',man_of_match)
  print('No. of boundries: ',boundaries)

```

```
# Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.
```

```
def venue_leader(venue):
  print(new[new['city']==venue].groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(4))
```

```
venue_leader('Delhi')
```
