## Q3. Make a function to find top bowler(in terms of most wickets) against any team. Take input as team.

```python
def Most_wickets(team):
  mask1=df2['dismissal_kind']=='caught'
  mask2=df2['dismissal_kind']=='bowled'
  mask3=df2['dismissal_kind']=='lbw'
  mask4=df2['dismissal_kind']=='stumped'
  mask5=df2['dismissal_kind']=='caught and bowled'
  mask6=df2['dismissal_kind']=='hit wicket'
  wicket=df2[mask1 | mask2 | mask3 | mask4 | mask5 | mask6]
  print(wicket[wicket['batting_team']==team].groupby('bowler').count().sort_values('player_dismissed',ascending=False).head(1))
```

```python
Most_wickets('Chennai Super Kings')
```

## Q4. Write a function which will take team name as input and return it's record of no. occurance in finals and no. of times won.  [Difficult]

ex.-- if team is CSK-- 
CSK in finals-

Played- 8
Won- 3

If any team which has not played in finals like Gujrat Lions

Gujrat Lions in finals-

Played-0


```python
def Finals(team): 
  final=df.drop_duplicates('season',keep='last')
  no_of_appearance=final[(final['team1']==team) | (final['team2']==team) ].count()[1]
  no_of_wins=final[(final['team1']==team) | (final['team2']==team) & (final['winner']==team)].count()[1]
  print('No of Final Appearance')
  print (no_of_appearance)
  print('No of Final Wins')
  print (no_of_wins)

```

```python
Finals('Chennai Super Kings')
```

## Q5. Write a function which will take a input as batsman and in output it gives total runs scored against each team. [Medium]




```python
 def batsmanVteam(team): 
  print(df2[df2['batsman']==team].groupby('bowling_team').sum().sort_values('batsman_runs',ascending=False))
```
